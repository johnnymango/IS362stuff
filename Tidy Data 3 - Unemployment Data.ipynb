# Johnny's Unemployment Data

```python
import pandas as pd
```

First, we import the unemployment data.

```python
unemploymentdata = pd.read_csv('https://raw.githubusercontent.com/johnnymango/IS362stuff/master/unemploymentdata.csv')
unemploymentdata.head()
```

To tidy this up, I first name the unamed column and rename the other columns to more convenient, shorter names.

```python
unemploymentdata.columns = ['Year', 'Less_HS', 'HS_Dipl', 'Some_Col', 'Assoc_Deg', 'Bach_Higher']
unemploymentdata.head()
```

Looking at the entire data set indicates that some of the data is missing and is represented by NaNs.

```python
unemploymentdata
```

To address the nans, I decided to replace them with the mean of the each column.  This will assist with later analysis.

```python
empdata = unemploymentdata.fillna(unemploymentdata.mean())
empdata
```
