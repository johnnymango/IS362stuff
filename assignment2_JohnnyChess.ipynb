```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""Chessmaster main chess class."""

import time

class ChessPiece(object):
    """ChessPiece definition"""

    prefix = ''

    def __init__(self, position):
        """Constructor for the ChessPiece() class.

        Args:
            position (integer): The notation of the current position in
                algebraic notation.

        Attributes:
            position (integer): The notation of the current position in
                algebraic notation.
            moves (list): The list moves in tuples.
        """
        self.moves = []
        self.position = position.lower()
        if not self.is_legal_move(self.position):
            message = '{} is not a legal start position'
            raise ValueError(message.format(position))

    def algebraic_to_numeric(self, tile):
        """Converts a string tile and converts it to a tuple of ints.

        Args:
            tile (string): The algebraic notation of a tile.

        Returns:
            An (int, int) tuple representing the tile or None if the the tile
            is invalid.

        Example:

            $ piece.algebraic_to_numeric('a1')
            $ (0, 0)

            $ piece.algebraic_to_numeric('h8')
            $ (7, 7)

            $ piece.algebraic_to_numeric('i9')
            $ None
        """
        if len(tile) == 2:
            lval = ord(tile[0]) - 97
            rval = int(tile[1]) - 1
            num_val = (lval, rval) if lval < 8 and rval < 8 else None
        else:
            num_val = None, None
        return num_val

    def is_legal_move(self, position):
        """Checks if position is valid.

        Args:
            position (integer): The notation of the position in algebraic
                notation.

        Returns:
            (boolean): True if the position is valid.

        Examples:

            $ piece.is_legal_move('a0')
            $ False

            $ piece.is_legal_move('')
            $ False

            $ piece.is_legal_move('h8')
            $ True
        """
        return isinstance(self.algebraic_to_numeric(position), tuple)

    def move(self, position):
        """Moves a chess piece.

        Args:
            position (integer): The notation of the position in algebraic
                notation.

        Returns:
            A tuple of the prefixed old and new positions and the timestamp or
            False if the move was invalid.
        """
        new_move = False
        if self.is_legal_move(position):
            new_move = (self.prefix + self.position,
                        self.prefix + position,
                        time.time())
            self.moves.append(new_move)
            self.position = position
        return new_move

class Rook(ChessPiece):
    """Rook definition"""

    prefix = 'R'

    def __init__(self, position):
        """Constructor for the Rook() class.

        Args:
            position (integer): The notation of the current position in
                algebraic notation.

        Attributes:
            position (integer): The notation of the current position in
                algebraic notation.
            moves (list): The list moves in tuples.
        """
        ChessPiece.__init__(self, position)

    def is_legal_move(self, position):
        """Checks if a rook's position is valid.

        A rook can only move along one access at a time.

        Args:
            position (integer): The notation of the position in algebraic
                notation.

        Returns:
            (boolean): True if the position is valid.

        Examples:

            $ left_rook.position
            $ 'a1'
            $ left_rook.is_legal_move('a5')
            $ True
            $ left_rook.is_legal_move('b2')
            $ False
        """
        is_valid = False
        position = position.lower()
        if self.algebraic_to_numeric(position) is not None:
            lval1, rval1 = self.algebraic_to_numeric(position)
            if self.position:
                lval2, rval2 = self.algebraic_to_numeric(self.position)
                is_valid = lval1 == lval2 or rval1 == rval2
            else:
                is_valid = True
        return is_valid
```
