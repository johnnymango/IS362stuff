## The Python Code to Get the JSON Data

The code uses the request module to pass the url string and the api-key.  I create a variable that converts the data to a json format.  The results of mydata match the NYT API results.

```python
import requests
import json
import pandas as pd

nytdata = requests.get("https://api.nytimes.com/svc/mostpopular/v2/mostviewed/Technology/30.json?&api-key=bf920907c8b54a7b9466f378e0f6a681")
mydata = nytdata.json()
mydata
```

## Convert the JSON to the Pandas dataframe.  

This code converts the json to a datafrome; it finds the data in the 'results' key of the dict.

```python
most_pop = pd.DataFrame(mydata['results'])
most_pop
```
