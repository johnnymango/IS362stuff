# Johnny's Auto Data Plots

First, we import all of the libraries we will need for the plots.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Import and Clean Up

We import the auto data and clean it up a bit by adding headers names.

```python
autodata = pd.read_csv('https://raw.githubusercontent.com/johnnymango/IS362stuff/master/autodata.csv', header=None)
autodata.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'name']
autodata.head()
```

To perform math on the columns, we need to replace the '?' with Nan's.  This will allow the horsepower column to be converted to a float. These are the current data types.

```python
autodata.dtypes
```

```python
autodata = autodata.replace('?', np.nan)
autodata['horsepower']=autodata.horsepower.astype(float)
autodata.dtypes
```

We also update the Origin columns to replace the numeric values with meaningful names: USA, Europe, Asia.

```python
autodata['origin'].replace(1, 'USA', inplace=True)
autodata['origin'].replace(2, 'Asia', inplace=True)
autodata['origin'].replace(3, 'Europe', inplace=True)
autodata.head()
```

## The Plots

This bar chart shows the distribution of cylinders.

```python
cyl_distr = autodata['cylinders'].plot.hist()
cyl_distr.set_xlabel('Cylinders')
plt.show()
```

This scatter plot shows the relationship between horsepower and weight.

```python
weight_HP_scatter = autodata.plot.scatter(x='weight', y='horsepower')
plt.show()
```

## The Seaborn Plots

This is series of seaborn plots with some insights into the data.

```python
#tips = sns.load_dataset("tips")
```

```python
sns.set(color_codes=True)
```

This plot shows, as expected, that as weight increases, mpg drops.  European cars have the worst mpg as weight increases while cars in the US and Asia are very similar.

```python
sns.lmplot(x="weight", y='mpg', hue='origin', data=autodata);
```

However, MPG have increased similarly in Europe and Asia.  US cars have also increased mpg over the time but are below non-US cars.

```python
sns.lmplot(x='model_year', y='mpg', hue='origin', data=autodata);
```

This plot indicates that for European and US cars, as weight increases, acceleration decreases as well.  Asia cars move in the opposite direction; Asian vehicle has greater acceleration as weight increases.

```python
sns.lmplot(x='weight', y='acceleration', hue='origin', data=autodata);
```

These side-by-side plots show that over time, horsepower has remained stable with slight decreases for European and Asian cars.  US cars greatly reduced horsepower to match horsepower of European and Asian cars.

```python
sns.lmplot(x='model_year', y='horsepower', col='origin', data=autodata);
```
